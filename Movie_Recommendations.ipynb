## The purpose of this notebook is to allow users to receive movie recommendations based on an inputted movie title

```python
# Reading in the movie data with pandas
import pandas as pd

movies = pd.read_csv(r"C:\Users\zareb\Downloads\movie_recs\ml-25m\movies.csv")
movies
```

```python
# Cleaning the movie titles with regex
import re

def clean_title(title):
    #Searching through each title and removing unnecessary characters in titles
    return re.sub("[^a-zA-Z0-9 ]", "", title) 

#Clean every movie title
movies["clean_title"] = movies["title"].apply(clean_title)
movies
```

```python
#Create the TFIDF Matrix for the search engine
#Convert the titles into sets of numbers for the search engine to work
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(ngram_range=(1,2))
tfidf = vectorizer.fit_transform(movies["clean_title"])
```

```python
#Create a search function 
#Need to compute the similarity between a term that we enter and the movies in the rest of the list
#Use cosine similarity

from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def search(title):
    title = clean_title(title)
    query_vector = vectorizer.transform([title])
    #Compare the query term with each of the titles in the dataset
    similarity = cosine_similarity(query_vector, tfidf).flatten()
    indices = np.argpartition(similarity, -5) [-5:]
    recommendations = movies.iloc[indices] [::-1]
    return recommendations
```

```python
# Build the interactive search box
import ipywidgets as widgets
from IPython.display import display

movie_input = widgets.Text(
    value="Toy Story",
    description="Movie Title:",
    disabled=False
)

movie_list = widgets.Output()

def when_type(data):
    with movie_list:
        movie_list.clear_output()  
        title = data["new"]  
        if len(title) > 5:  
            results = search(title)  
            display(results)  

movie_input.observe(when_type, names='value')
display(movie_input, movie_list)
```

```python
#Read in the movie ratings data
ratings = pd.read_csv(r"C:\Users\zareb\Downloads\movie_recs\ml-25m\ratings.csv")
ratings
```

```python
#Toy Story Movie ID
movie_id = 1
```

```python
#Now we want to use the ratings as a metric for the recommendations
similar_users = ratings[(ratings["movieId"] == movie_id) & (ratings["rating"] > 4)]["userId"].unique()

similar_user_recs = ratings[(ratings["userId"].isin(similar_users)) & (ratings["rating"] > 4)]["movieId"]

similar_user_recs = similar_user_recs.value_counts() / len(similar_users)
similar_user_recs = similar_user_recs[similar_user_recs > .10]

#How much all the users like movies
all_users = ratings[(ratings["movieId"].isin(similar_user_recs.index)) & (ratings["rating"] > 4)]
all_user_recs = all_users["movieId"].value_counts() / len(all_users["userId"].unique())
```

```python
#Create a recommendation score system
rec_percentages = pd.concat([similar_user_recs, all_user_recs], axis=1)
rec_percentages.columns = ["similar", "all"]

rec_percentages
```

```python
#Finding the score of movies and using it as a recommendation systenm
rec_percentages["score"] = rec_percentages["similar"] / rec_percentages["all"]
rec_percentages = rec_percentages.sort_values("score", ascending=False)
rec_percentages.head(10).merge(movies, left_index=True, right_on="movieId")
```

```python
#Building the recommendation function
def find_similar_movies(movie_id):
    similar_users = ratings[(ratings["movieId"] == movie_id) & (ratings["rating"] > 4)]["userId"].unique()
    similar_user_recs = ratings[(ratings["userId"].isin(similar_users)) & (ratings["rating"] > 4)]["movieId"]
    similar_user_recs = similar_user_recs.value_counts() / len(similar_users)

    similar_user_recs = similar_user_recs[similar_user_recs > .10]
    all_users = ratings[(ratings["movieId"].isin(similar_user_recs.index)) & (ratings["rating"] > 4)]
    all_user_recs = all_users["movieId"].value_counts() / len(all_users["userId"].unique())
    rec_percentages = pd.concat([similar_user_recs, all_user_recs], axis=1)
    rec_percentages.columns = ["similar", "all"]
    
    rec_percentages["score"] = rec_percentages["similar"] / rec_percentages["all"]
    rec_percentages = rec_percentages.sort_values("score", ascending=False)
    return rec_percentages.head(10).merge(movies, left_index=True, right_on="movieId")[["score", "title", "genres"]]
```

```python
#Create the widget which takes in a movie and outputs recommendations
import ipywidgets as widgets
from IPython.display import display

movie_name_input = widgets.Text(
    value='Toy Story',
    description='Movie Title:',
    disabled=False
)
recommendation_list = widgets.Output()

def on_type(data):
    with recommendation_list:
        recommendation_list.clear_output()
        title = data["new"]
        if len(title) > 5:
            results = search(title)
            movie_id = results.iloc[0]["movieId"]
            display(find_similar_movies(movie_id))

movie_name_input.observe(on_type, names='value')

display(movie_name_input, recommendation_list)
```
